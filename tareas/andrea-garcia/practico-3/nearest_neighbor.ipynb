```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Invistico_Airline.csv')
```

```python
df.satisfaction.unique()
```

```python
df.satisfaction = df.satisfaction.map({'satisfied': 1, 'dissatisfied': 2})
```

```python
df['Customer Type'].unique()
```

```python
df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 2})
```

```python
df['Type of Travel'].unique()
```

```python
df['Type of Travel'] = df['Type of Travel'].map({'Personal Travel': 1, 'Business travel': 2})
```

```python
df.Class.unique()
```

```python
df.Class = df.Class.map({'Eco': 1, 'Business': 2, 'Eco Plus': 3})
```

```python
df.dtypes
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean())
```

```python
df.describe()
```

```python
corr = round(df.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = df.drop(['satisfaction', 'Departure/Arrival time convenient', 'Gate location', 'Arrival Delay in Minutes', 'Flight Distance'], axis='columns')
data_y = df.satisfaction
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.10)
```

```python
k_neighbors = KNeighborsClassifier(n_neighbors=6)
```

```python
k_neighbors.fit(train_x, train_y)
```

```python
y_predicted = k_neighbors.predict(test_x)
```

```python
print(f'Score: {round(k_neighbors.score(test_x, test_y),4) * 100}%')
```

```python
customer_type = ['None', 'Leal', 'Desleal']
type_travel = ['None', 'Viaje Personal', 'Viaje de Negocios']
clase = ['None', 'Eco', 'Business', 'Eco Plus']
state = ['insatisfecho', 'satisfecho']

def passenger_satisfaction_classifier(model, customer, age, travel, p_class, sit_comfort, food_and_drink, wifi, entertaiment, online_support, online_booking, service, leg_room, baggage_handling, checkin, cleanliness, online_boarding, departure_delay):
    predicted = model.predict([[customer, age, travel, p_class, sit_comfort, food_and_drink, wifi, entertaiment, online_support, online_booking, service, leg_room, baggage_handling, checkin, cleanliness, online_boarding, departure_delay]])
    print(f'El pasajero de tipo: {customer_type[customer]}, de {age} años, que hizo un {type_travel[travel]}, en clase {clase[p_class]}, y con calificaciones de comodidad: {sit_comfort}, comida y bebida: {food_and_drink}, wifi: {wifi}, entretenimiento: {entertaiment}, soporte en línea: {online_support}, reserva en línea: {online_booking}, servicio a bordo: {service}, espacio para las piernas: {leg_room}, manejo de equipaje: {baggage_handling}, check-in: {checkin}, limpieza: {cleanliness}, abordaje en línea: {online_boarding}, y demora en la salida de: {departure_delay} minutos, está: {state[predicted[0]]} con su experiencia de vuelo.')

```
