```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import tree
```

```python
df = pd.read_csv('bank.csv')
```

```python
df.head()
```

```python
df.drop([ 'job','marital','education','default', 'housing',
'loan', 'contact', 'month', 'poutcome',] ,axis=1 , inplace=True)
```

```python
df.describe()
```

```python
df['deposit'] = df.deposit.map({'yes':0, 'no':1})
```

```python
df.describe()
```

```python
data_x = df.drop('balance', axis='columns')
data_y = df.deposit
```

```python
data_x
```

```python
decission_model = DecisionTreeClassifier()
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, train_size=0.2)
```

```python
decission_model.fit(train_x, train_y)
```

```python
y_predicted = decission_model.predict(test_x)
```

```python
print(f'Score del modelo: {round(decission_model.score(test_x, test_y), 4) * 100}%')
```

```python
tree.plot_tree(decission_model)
plt.show()
```
