```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['blood_glucose_level'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
df.smoking_history.unique()
```

```python
df.smoking_history = df.smoking_history.map({'never':1, 'No Info':2, 'current':3, 'former':4, 'ever':5, 'not current':6})
```

```python
df.gender = df.gender.map({'Female': 0 , 'Male': 1})
```

```python
df.head()  
```

```python
corr = round(df.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = df.drop(['diabetes'], axis='columns')
data_y = df.diabetes
```

```python
data_x
```

```python
data_y
```

```python
df.describe()
```

```python
data_x.gender = data_x.gender.fillna(data_x.gender.mean())
```

```python
# Ahora inicializamos el modelo
decission_model = DecisionTreeClassifier()
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
# Entrenamos el modelo
decission_model.fit(train_x, train_y)
```

```python
y_predicted = decission_model.predict(test_x)
```

```python
print(f'Score del modelo: {round(decission_model.score(test_x, test_y), 4) * 100}%')
```

```python
result = ['No tiene diabetes', 'Tiene diabetes']
gender = {0: 'None', 1: 'Masculino', 2: 'Femenino'}
```

```python
def classify_diabetes(model, gender, age, bmi, bp, smoking_history, avg_glucose_level):
    predicted = model.predict([gender, age, bmi, bp, smoking_history, avg_glucose_level])
    print(f'El paciente de género {gender[gender]}, de {age} años, con un BMI de {bmi}, presión arterial de {bp}, historial de tabaquismo {smoking_history}, nivel de glucosa promedio {avg_glucose_level}: {result[predicted[0]]}')
```

```python
classify_diabetes(decission_model, gender=1, age=50, bmi=30, bp=100, smoking_history=1, avg_glucose_level=120)
```
