```python
from sklearn.ensemble import RandomForestClassifier

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
## cargando el dataset
df = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['blood_glucose_level'], axis='columns', inplace=True)
```

```python
df.gender.unique()
```

```python
df.smoking_history.unique()
```

```python
df.gender = df.gender.map({'Female': 0 , 'Male': 1, 'Other': 2})
```

```python
df.smoking_history = df.smoking_history.map({'never':1, 'No Info':2, 'current':3, 'former':4, 'ever':5, 'not current':6})
```

```python
df.head()
```

```python
corr = round(df.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = df.drop(['diabetes'], axis='columns')
data_y = df.diabetes
```

```python
forest = RandomForestClassifier()
```

```python
forest.fit(data_x, data_y)
```

```python
print(f'El score del modelo es: {round(forest.score(data_x, data_y),4)*100}%')
```
