```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
```

```python
#dataset de fumadores y consumidores de alcohol
df = pd.read_csv('smok_drk.csv')
```

```python
df.DRK_YN = df.DRK_YN.map({'Y':1,'N':2})
df.sex	 = df.sex	.map({'Male':1,'Female':2})
```

```python
df.drop([ 'height','sight_left','sight_left','sight_right','hear_left','hear_right','urine_protein','serum_creatinine','SGOT_AST','SGOT_ALT' ], axis=1, inplace=True)
```

```python
corr = round(df.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = df.drop(['SMK_stat_type_cd', 'DRK_YN'], axis='columns')  
data_y_smoker = df.SMK_stat_type_cd  
data_y_alcohol = df.DRK_YN  
```

```python
train_x, test_x, train_y_smoker, test_y_smoker, train_y_alcohol, test_y_alcohol = train_test_split(data_x, data_y_smoker, data_y_alcohol, test_size=0.1)
```

```python
k_neighbors = KNeighborsClassifier()

k_neighbors_smoker = KNeighborsClassifier()

k_neighbors_alcohol = KNeighborsClassifier()
```

```python
k_neighbors_smoker.fit(train_x, train_y_smoker)

k_neighbors_alcohol.fit(train_x, train_y_alcohol)
```

```python
score_smoker = k_neighbors_smoker.score(test_x, test_y_smoker)
score_alcohol = k_neighbors_alcohol.score(test_x, test_y_alcohol)
```

```python
print(f'Precisión del modelo de fumadores: {round(score_smoker * 100, 2)}%')
print(f'Precisión del modelo de consumo de alcohol: {round(score_alcohol * 100, 2)}%')
```

```python
result = [ 'bebe_alcohol', 'no_bebe_alcohol' ]

def classify(model, sex,age, weight,waistline,SBP,DBP,BLDS, tot_chole, HDL_chole, LDL_chole,triglyceride, hemoglobin, gamma_GTP):
    prediccion_alcohol = k_neighbors_alcohol.predict([[sex, age, weight, waistline, SBP, DBP, BLDS, tot_chole, HDL_chole, LDL_chole, triglyceride, hemoglobin, gamma_GTP]])
    print(f'Predicción del consumo de alcohol: {result[prediccion_alcohol[0]]}')
```
