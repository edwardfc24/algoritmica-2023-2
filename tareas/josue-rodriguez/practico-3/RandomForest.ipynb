```python
from sklearn.ensemble import RandomForestClassifier

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('smok_drk.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.DRK_YN = df.DRK_YN.map({'Y':1,'N':2})
df.sex	 = df.sex	.map({'Male':1,'Female':2})
```

```python
df.drop([ 'height','sight_left','sight_left','sight_right','hear_left','hear_right','urine_protein','serum_creatinine','SGOT_AST','SGOT_ALT' ], axis=1, inplace=True)
```

```python
data_x = df.drop(['SMK_stat_type_cd', 'DRK_YN'], axis='columns')  
data_y = df.DRK_YN  
```

```python
data_x.describe()
```

```python
data_x
```

```python
forest = RandomForestClassifier()
```

```python
forest.fit(data_x, data_y)
```

```python
print(f'El score del modelo es: {round(forest.score(data_x, data_y), 4) *100}%')
```

```python
result = [ 'bebe_alcohol', 'no_bebe_alcohol' ]

def drink_classify(model, sex,age, weight,waistline,SBP,DBP,BLDS, tot_chole, HDL_chole, LDL_chole,triglyceride, hemoglobin, gamma_GTP):
    prediccion_alcohol = forest.predict([[sex, age, weight, waistline, SBP, DBP, BLDS, tot_chole, HDL_chole, LDL_chole, triglyceride, hemoglobin, gamma_GTP]])
    print(f'Predicción del consumo de alcohol: {result[prediccion_alcohol[0]]}')
```

```python
drink_classify(forest, 1, 40, 70, 80, 120, 80, 90, 180, 50, 100, 70, 14, 30)
```
