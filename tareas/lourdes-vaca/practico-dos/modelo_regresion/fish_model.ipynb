```python
data_x = df.drop(['Species', 'Weight'], axis=1)
data_y = df.Weight
```

```python
data_x
```

```python
data_y
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.2)
```

```python
#comprueba el conjunto de entrenamiento (train) y la muestra/prueba (test)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
#observo que contiene train_x
train_x
```

```python
#observo que contiene train_y
train_y
```

```python
#inicializo el modelo a entrenar
fish_reg = LinearRegression()
```

```python
#entreno el modelo
fish_reg.fit(train_x, train_y)
```

```python
#clasificaciÃ³n de los datos
y_predicted = fish_reg.predict(test_x)
```

```python
#observo que contiene y_predicted
y_predicted
```

```python
print(f'CalificaciÃ³n de entrenamiento: {round(fish_reg.score(train_x, train_y), 4)*100}%')
print(f'CalificaciÃ³n de predicciÃ³n: {round(fish_reg.score(test_x, test_y), 4)*100}%')
```
