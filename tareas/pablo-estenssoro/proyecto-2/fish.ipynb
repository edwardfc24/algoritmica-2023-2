```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("dataset_fish.csv")
```

```python
# Verificamos data frame
df.head(76)
```

```python
df.info()
```

```python
df.Weight.unique()
```

```python
#mapeamos
df.Species = df.Species.map({"Bream":1,"Roach":2,"Whitefish":3,"Parkki":4,"Perch":5,"Pike":6,"Smelt":7})
```

```python
df.describe()
```

```python
corr = round(df.corr(),4)
corr.style.background_gradient()
```

```python
x = df.drop(["Species","Weight"] ,axis=1)
y = df.Weight
```

```python
x
```

```python
train_x,test_x,train_y,test_y  = train_test_split(x,y,test_size=0.20)
```

```python
#iniciamos el modelo
linear = LinearRegression()
```

```python
#entrenamos el modelo
linear.fit(train_x,train_y)
```

```python
y_predicted = linear.predict(test_x)
```

```python
print(f"calificacion de prediccion {linear.score(test_x,test_y)}")
```
