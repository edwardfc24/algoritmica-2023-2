**Ridge es una variante de regresión lineal también conocida como "Ridge regression", para mejorar la generalización del modelo y reducir el riesgo de sobreajuste en problemas de regresión lineal múltiple.**


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

import pandas as pd
import numpy as np
```

```python
#Importamos el dataset y lo convertimos en un DataFrame
df = pd.read_csv("dataset_fish.csv")
```

```python
df.head()
```

```python
df.Species.unique()
```

```python
#mapeamos columna de tipo Object categorica especies
df.Species = df.Species.map({"Bream":1,"Roach":2,"Whitefish":3,"Parkki":4,"Perch":5,"Pike":6,"Smelt":7})
```

```python
df.tail()
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
#Inicializamos el modelo
ridge = Ridge()
```

```python
#preparamos la informacion
data_x = df.drop(["Weight"],axis=1)
data_y = df.Weight
```

```python
train_x,test_x,train_y,test_y = train_test_split(data_x,data_y,test_size=0.25)
```

```python
#Entrenamos el modelo
ridge.fit(train_x,train_y)
```

```python
#clasificacion
print(f"Score for prediction : {round(ridge.score(test_x,test_y),4)*100} %")
```

```python
#predecir
all =["None","Bream","Roach","Whitefish","Parkki","Perch","Pike","Smelt"]
species = 1
length1 = 24
length2 = 18
length3 = 22
height = 11
width = 4
predict = np.round(ridge.predict([[species,length1,length2,length3,height,width]]),2)
```

```python
print(f"El pez de la especie {all[2]} con lenght1 de {length1},length2 de {length2},length3 de{length3},con un height de {height} y un width de {width} tiene un peso de {predict}")
```
