```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('riceClassification.csv')
```

```python
df.head()
```

```python
df.drop(['id', 'AspectRation'], axis='columns', inplace=True)
```

```python
df.Class.unique()
```

```python
df
```

```python
corr = round(df.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = df.drop(['Class'], axis='columns')
data_y = df.Class
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
k_neighbors = KNeighborsClassifier()
```

```python
k_neighbors.fit(train_x, train_y)
```

```python
y_predicted = k_neighbors.predict(test_x)
```

```python
print(f'Score: {round(k_neighbors.score(test_x, test_y), 4) * 100}%')
```

```python
clasificacion = ['bueno', 'malo']

Area = 1000.0
MajorAxisLength = 13.2
MinorAxisLength = 11.8
Eccentricity = 10.6
ConvexArea = 19.5
EquivDiameter = 12.5
Extent = 10.5
Perimeter = 11.5
Roundness = 10.5
```

```python
def rice_classify(model, Area, MajorAxisLength, MinorAxisLength, Eccentricity, ConvexArea, EquivDiameter, Extent, Perimeter, Roundness):
    predicted =  model.predict([[Area, MajorAxisLength, MinorAxisLength, Eccentricity, ConvexArea, EquivDiameter, Extent, Perimeter, Roundness]])
    print(f'El arroz de area: {Area}, de mayor longitud: {MajorAxisLength} y de menor longitud: {MinorAxisLength}, y de exentricidad: {Eccentricity} y de area convexa {ConvexArea} que tiene un diametro de: {EquivDiameter} con un perimetro de: {Perimeter} entrenamiento, con puntuacion de redondez: {Roundness} sera {clasificacion[predicted[0]]}')
    
```

```python
rice_classify(k_neighbors, Area, MajorAxisLength, MinorAxisLength, Eccentricity, ConvexArea, EquivDiameter, Extent, Perimeter, Roundness)
```
